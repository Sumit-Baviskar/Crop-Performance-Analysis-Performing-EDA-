- The problem is to analyze agricultural crop yield data to identify key factors influencing productivity, detect anomalies, and uncover patterns that can inform strategies for optimizing crop yields and sustainable farming practices.

## **Introduction to Exploratory Data Analysis (EDA) on Agriculture Crop Yield :**

- **Exploratory Data Analysis (EDA)** is a fundamental step in any data analysis process that aims to gain a deep understanding of a dataset’s structure, identify key patterns, and uncover relationships among variables. In the context of agriculture, EDA becomes particularly important as it provides critical insights into the factors influencing crop yield, enabling farmers, researchers, and policymakers to make informed decisions. Agriculture is inherently complex, involving numerous variables such as environmental conditions, soil health, farming techniques, and climatic factors, all of which have a significant impact on crop productivity.

- The **process of EDA helps in transforming raw agricultural data into actionable knowledge** by applying a variety of statistical and visualization techniques. This phase involves the exploration of different aspects of the dataset, including the distribution of key variables, the detection of outliers and anomalies, handling missing data, and identifying underlying trends or patterns that could guide future analysis or decision-making.

- Agricultural productivity is **influenced by a multitude of factors, many of which are interrelated**. By conducting a thorough EDA on agricultural crop yield data, it becomes possible to uncover hidden relationships between these variables and identify the most influential factors that drive productivity. For example, how do soil quality, temperature, and rainfall interact to affect crop yield? Can specific farming techniques be correlated with improved crop outputs? What impact do regional or seasonal variations have on the yield of different crops? Answering these questions through EDA will help establish the foundation for more sophisticated modeling, predictive analytics, and optimization strategies aimed at improving crop yields and promoting sustainable agricultural practices.

- EDA is **not only valuable for identifying patterns but also serves as a powerful tool for detecting issues such as data errors or inconsistencies, such as outliers, missing values, or incorrect entries**. These anomalies can be critical in shaping the approach for subsequent data cleaning and preparation, which is an essential part of the data analysis pipeline.

- Ultimately, the insights derived from EDA in the agriculture crop yield dataset can provide valuable guidance for enhancing farming practices, ensuring food security, and fostering a more efficient, environmentally friendly approach to agriculture. The findings from this initial exploratory analysis will guide more targeted investigations, hypothesis testing, and predictive modeling in future stages of the project. Through a **systematic examination of the data, the EDA process will contribute to optimizing agricultural practices, improving yields, and reducing resource wastage, leading to better outcomes for farmers and the agricultural industry as a whole**.








## **Purpose of EDA on Agriculture Crop Yield :**
The **purpose of performing EDA** on agricultural crop yield data is to:

- **Understand the structure of the data :** Gain an initial understanding of how the data is organized and what variables are present, such as crop type, yield, soil conditions, and environmental factors.

- **Detect outliers and anomalies :** Identify extreme or unusual data points that could distort the analysis or suggest errors in data collection.

- **Identify missing values :** Highlight any missing or incomplete data that could affect the results of the analysis and suggest strategies for handling these gaps.
- **Find important variables and patterns :** Investigate which factors (e.g., climate conditions, soil types, or farming techniques) have the most influence on crop yields, and identify trends or correlations that could inform future decision-making.

#**About Dataset**

This dataset contains agricultural data for **1,000,000 samples** aimed at predicting crop yield (in tons per hectare) based on various factors.




- **Region :** The geographical region where the crop is grown (North, East, South, West).




- **Soil_Type :** The type of soil in which the crop is planted (Clay, Sandy, Loam, Silt, Peaty, Chalky).



- **Crop :** The type of crop grown (Wheat, Rice, Maize, Barley, Soybean, Cotton).



- **Rainfall_mm :** The amount of rainfall received in millimeters during the crop growth period.



- **Temperature_Celsius :** The average temperature during the crop growth period, measured in degrees Celsius.


- **Fertilizer_Used :** Indicates whether fertilizer was applied (True = Yes, False = No).



- **Irrigation_Used :** Indicates whether irrigation was used during the crop growth period (True = Yes, False = No).



- **Weather_Condition :** The predominant weather condition during the growing season (Sunny, Rainy, Cloudy).



- **Days_to_Harvest :** The number of days taken for the crop to be harvested after planting.



- **Yield_tons_per_hectare :** The total crop yield produced, measured in tons per hectare.

##**Treating Outliers**

```python
# Using IQR(Interquartile range)
# Percentile
# IQR=Q3-Q1 (range of middle 50% of the data)

Q1=df.describe()["Yield_tons_per_hectare"]["25%"]
print(f"Q1 = {Q1}")
Q3=df.describe()["Yield_tons_per_hectare"]["75%"]
print(f"Q3 = {Q3}")
```

```python
IQR=Q3-Q1
print(f"IQR = {IQR}")
```

```python
Lower_limit=Q1-1.5*IQR
Upper_limit=Q3+1.5*IQR

print(f"Lower_limit = {Lower_limit}")
print(f"Upper_limit = {Upper_limit}")
```

###2. **Densest Areas Indicate Ideal Rainfall:**

   - With fertilizer, yields between **5 and 7 tons** dominate this range, while without fertilizer, the peak yield remains slightly lower, around **4 to 6 tons/hectare**.

###3. **Yield at Extreme Rainfall Levels:**

   - For **rainfall levels below 300 mm**, yields are consistently lower, typically below **4 tons/hectare**, in both plots. This suggests that **insufficient rainfall** significantly limits crop yield.


   - At the higher end (above **900 mm** of rainfall), yields also tend to taper off, indicating that **excessive rainfall** may negatively impact crop productivity, even with the use of fertilizer.
   


#**Insights:**

- **Rainfall is a Strong Predictor of Yield:** A clear positive relationship is evident between rainfall and yield in both conditions (with and without fertilizer).


- **Fertilizer Enhances Yield:** The use of fertilizer boosts the yield across all rainfall levels, with the greatest benefit seen between **500-900 mm** of rainfall, where yields of **6-8 tons/hectare** are most common.


- **Optimal Rainfall Range for High Yields:** The ideal rainfall for maximum crop productivity lies between **500 mm to 700 mm**, beyond which yields tend to plateau or decline.


- **Low and High Rainfall Affect Yield Negatively:** Insufficient rainfall (<300 mm) leads to low yields, and excessive rainfall (>900 mm) also reduces yield, likely due to waterlogging or nutrient leaching.
